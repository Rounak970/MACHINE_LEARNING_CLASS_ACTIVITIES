```python
# APPLICATION OF GRADIENT DESCENT 

import numpy as np

x=np.array([1,2,3,4,5],float)
y=np.array([5,6,10,13,11],float)
lr=0.1

def mse(y,yh): return np.mean((yh-y)**2)

def batch(x,y,b0,b1,lr,ep=10):
    print("BATCH GD")
    for i in range(ep):
        yh=b0+b1*x
        e=mse(y,yh)
        db0=np.mean(yh-y)
        db1=np.mean((yh-y)*x)
        b0-=lr*db0; b1-=lr*db1
        print(i+1,": b0=",round(b0,4)," b1=",round(b1,4)," err=",round(e,4))
    print()
        
def sgd(x,y,b0,b1,lr,ep=10):
    print("SGD")
    for i in range(ep):
        for xi,yi in zip(x,y):
            yh=b0+b1*xi
            b0-=lr*(yh-yi)
            b1-=lr*(yh-yi)*xi
        e=mse(y,b0+b1*x)
        print(i+1,": b0=",round(b0,4)," b1=",round(b1,4)," err=",round(e,4))

batch(x,y,3.3,1.9,lr)
sgd(x,y,3.3,1.9,lr)
```
